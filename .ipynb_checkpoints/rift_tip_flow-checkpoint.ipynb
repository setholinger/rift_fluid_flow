```python
import numpy as np
from scipy.integrate import odeint
from scipy.optimize import fmin
import matplotlib.pyplot as plt
```

```python
'''

Define useful functions


'''

def stress_intensity(L,sigma):
    return sigma*np.sqrt(np.pi*L)

def find_min_rxx(L,sigma0,K_ic):
    # Find Rxx such that K_i = K_ic assuming no dynamic contribution from water level
    goal = lambda R : np.abs(K_ic - stress_intensity(L0,R+sigma0))
    Rxx_best = fmin(goal, 10e3)
    return Rxx_best
```

```python
'''

Define the model

'''

# this version does not account for alpha's dependence on time and assumes that w' = aL'
def water_crack(y, t, epsilon, H_i, H_c, Rxx):
    
    # get variables and set parameters
    eta, d_eta_dt, L = y
    g = 9.8
    K_c = 1e5
    rho_i = 910
    rho_w = 1000
    v_r = 2000
    
    # calculate water level
    H_w = rho_i/rho_w * H_i

    # calculate stress and stress intensity factor
    sigma = rho_i*g*H_i/2 - rho_w*g/(2*H_i)*eta**2
    sigma_sum = Rxx - sigma
    K_squared = np.pi*L*sigma_sum**2
    
    # use stress intensity factor to determine if crack propagates
    if K_squared < K_c**2:
        dLdt = 0
    else:
        dLdt = v_r * (1 - K_c**2 / K_squared)

    # use "Eshelby" relation to get width and its time derivative as a function of length
    mu_star = 3.6e9/(1-0.33)
    alpha = sigma_sum / mu_star * np.pi/4
    w = alpha * L
        
    # write time derivative of water volume, V'
    dVdt = w*L*d_eta_dt + 2*w*dLdt*eta
    
    # write time derivative of stress
    d_sigma_sum_dt = (rho_w*g)/H_i * d_eta_dt * eta
    
    # write 2nd time derivative of crack length
    dLd2t = (v_r*K_c**2)/np.pi * (1/(L*sigma_sum**2))*(dLdt/L + 2*d_sigma_sum_dt/sigma_sum)
    
    # write expression for terms of V" that do not contain eta"
    lamb = 2*w*dLd2t*eta + 3*w*dLdt*d_eta_dt + dLdt*alpha*(L*d_eta_dt + 2*dLdt*eta)
    
    # write bernoulli with u' substituted in so all eta" terms can be moved to LHS
    #d_eta_d2t = (1/(H_w+epsilon*w)) * (g*eta + (epsilon/L**2)*(dVdt*dLdt - lamb*L)) # steph's linearized bernoulli
    #d_eta_d2t = (1/(H_w+epsilon*w/H_c)) * (-2*g*(eta-H_w) + (epsilon/H_c/L**2)*(dVdt*dLdt - lamb*L)) # brad's linearized bernoulli
    d_eta_d2t = (1/(epsilon*w+eta)) * ((epsilon/L**2)*(dVdt*dLdt - lamb*L) - 0.5*(d_eta_dt**2 + (1/L**2/H_c**2)*dVdt**2) + g*(H_w - eta)) # nonlinear bernoulli
    
    dydt = [d_eta_dt, d_eta_d2t, dLdt]
    
    return dydt



'''

Define parameters

'''

H_i = 200
rho_i = 910
rho_w = 1000
H_w = rho_i/rho_w * H_i
g = 9.8
eta0 = H_w
d_eta_dt0 = 0
L0 = 5e3
w0 = 10
K_ic = 1e5
sigma0 = -rho_i*g*H_i/2 + rho_w*g/(2*H_i)*eta0**2

# The following calculates Rxx such that K_ic = K_i, and then adds a tiny perturbation
Rxx = find_min_rxx(L0,sigma0,K_ic)[0]+1



'''

Run simulations

'''

y0 = [eta0, d_eta_dt0,L0]
t = np.linspace(0, 750, 10001)

# epsilon = 10e3/400 #"Lc/Hc"

fig,ax=plt.subplots(3,1,figsize=(10,10))
fig.patch.set_facecolor('w')

for epsilon in (25,100,400):

    H_c = 20e3/epsilon
    sol = odeint(water_crack, y0, t, args=(epsilon, H_i, H_c, Rxx))

    ax[0].plot(t, sol[:, 0]-H_w)
    ax[0].set_ylabel('Deflection of water surface (m)')
    ax[1].plot(t, np.gradient(sol[:, 2],t), label='L(t), km')
    ax[1].set_ylabel('Crack propagation rate (m/s)')
    ax[2].plot(t, sol[:, 2]/1e3-L0/1e3, label='L(t), km')
    ax[2].set_ylabel('Crack growth (km)')
    print("Mean crack propagation rate: ",np.round(np.mean(np.gradient(sol[:, 2],t)),2)," m/s")
ax[0].grid()
ax[1].grid()
ax[2].grid()
plt.xlabel('time, seconds')
plt.show()
```

```python
'''

Define model

'''

# this version does account for alpha's dependence on time and assumes that w' = aL' + a'L
def water_crack(y, t, epsilon, H_i, H_c, Rxx):
    
    # get variables and set parameters
    eta, d_eta_dt, L = y
    g = 9.8
    K_c = 1e5
    rho_i = 910
    rho_w = 1000
    v_r = 2000
    
    # calculate water level
    H_w = rho_i/rho_w * H_i

    # calculate stress and stress intensity factor
    sigma = rho_i*g*H_i/2 - rho_w*g/(2*H_i)*eta**2
    sigma_sum = Rxx - sigma
    K_squared = np.pi*L*sigma_sum**2
    
    # use stress intensity factor to determine if crack propagates
    if K_squared < K_c**2:
        dLdt = 0
    else:
        dLdt = v_r * (1 - K_c**2 / K_squared)

    # use "Eshelby" relation to get width and its time derivative as a function of length
    mu_star = 3.6e9/(1-0.33)
    alpha = sigma_sum / mu_star * np.pi/4
    w = alpha * L

    # write time derivative of stress sum
    d_sigma_sum_dt = d_eta_dt*eta*rho_w*g/H_i
    
    # write time derivative of alpha
    d_alpha_dt = d_sigma_sum_dt*np.pi/mu_star/4
    
    # write time derivative of width
    dwdt = alpha*dLdt + d_alpha_dt*L
        
    # write time derivative of water volume, V'
    dVdt = w*L*d_eta_dt + w*dLdt*eta + dwdt*L*eta
    
    # write flow speed
    u = dVdt/L/H_c
    
    # write time derivative of stress
    d_sigma_sum_dt = (rho_w*g)/H_i * d_eta_dt * eta
    
    # write 2nd time derivative of crack length
    dLd2t = (v_r*K_c**2)/np.pi * (1/(L*sigma_sum**2))*(dLdt/L + 2*d_sigma_sum_dt/sigma_sum)
    
    # write expression for terms of V" that do not contain eta"
    gamma = rho_w*g*np.pi/4/H_i/mu_star
    lamb = 2*w*dLdt*d_eta_dt + 2*dwdt*L*d_eta_dt + 2*dwdt*dLdt*eta + w*dLd2t*eta + L*eta*(alpha*dLd2t + 2*d_alpha_dt*dLdt + gamma*L*d_eta_dt**2)
    beta = L*(w + gamma*L*eta**2)
    
    # write bernoulli with u' substituted in so all eta" terms can be moved to LHS
    d_eta_d2t = (epsilon*beta/L+eta)**-1 * ((epsilon/L**2)*(dVdt*dLdt - lamb*L) - 0.5*(d_eta_dt**2 + u**2) + g*(H_w - eta)) # nonlinear bernoulli
    
    dydt = [d_eta_dt, d_eta_d2t, dLdt]
    
    return dydt


'''

Define parameters

'''

H_i = 200
rho_i = 910
rho_w = 1000
H_w = rho_i/rho_w * H_i
g = 9.8
eta0 = H_w
d_eta_dt0 = 0
L0 = 5e3
w0 = 10
K_ic = 1e5
sigma0 = -rho_i*g*H_i/2 + rho_w*g/(2*H_i)*eta0**2

# The following calculates Rxx such that K_ic = K_i, and then adds a tiny perturbation
Rxx = find_min_rxx(L0,sigma0,K_ic)[0]+1



'''

Run simulations

'''

y0 = [eta0, d_eta_dt0,L0]
t = np.linspace(0, 750, 10001)

# epsilon = 10e3/400 #"Lc/Hc"

fig,ax=plt.subplots(3,1,figsize=(10,10))
fig.patch.set_facecolor('w')

for epsilon in (25,100,400):

    H_c = 20e3/epsilon
    sol = odeint(water_crack, y0, t, args=(epsilon, H_i, H_c, Rxx))

    ax[0].plot(t, sol[:, 0]-H_w)
    ax[0].set_ylabel('Deflection of water surface (m)')
    ax[1].plot(t, np.gradient(sol[:, 2],t), label='L(t), km')
    ax[1].set_ylabel('Crack propagation rate (m/s)')
    ax[2].plot(t, sol[:, 2]/1e3-L0/1e3, label='L(t), km')
    ax[2].set_ylabel('Crack growth (km)')
    print("Mean crack propagation rate: ",np.round(np.mean(np.gradient(sol[:, 2],t)),2)," m/s")
ax[0].grid()
ax[1].grid()
ax[2].grid()
plt.xlabel('time, seconds')
plt.show()
```

```python
'''

Define model

'''

def water_crack(y, t, epsilon, H_i, Rxx, w):
    
    # get parameters and variables
    eta, d_eta_dt, L = y
    g = 9.8
    v_r = 2000
    K_ic = 1e5
    rho_i = 910
    rho_w = 1000
    H_w = rho_i/rho_w * H_i

    # calculate sum of stresses and stress intensity factor
    ice_pressure = rho_i * g * H_i / 2
    water_pressure = rho_w * g / (2*H_i) * eta**2
    sigma_sum =  Rxx - ice_pressure +  water_pressure
    d_sigma_sum_dt = d_eta_dt*eta*rho_w*g/H_i
    K_i = stress_intensity(L,sigma_sum)
    
    # calculate crack growth rate and its derivative
    if K_i < K_ic:
        dLdt = 0
    else:
        dLdt = v_r * (1 - (K_ic/K_i)**2)
    dLd2t = (v_r*K_ic**2)/(np.pi*(L**2)*(sigma_sum**3))*(2*L*d_sigma_sum_dt + sigma_sum*dLdt)
        
    # write water volume and its derivative
    V = w*L*eta
    dVdt = w*(dLdt*eta + L*d_eta_dt)
        
    # bernoulli equation- what sign should g have?
    d_eta_d2t = (-g*(eta-H_w) - (epsilon/L**2)*(L*w*(dLd2t*eta + 2*dLdt*d_eta_dt)-dVdt*dLdt)) / (H_w + w*epsilon)    

    dydt = [d_eta_dt, d_eta_d2t, dLdt]
    
    return dydt



'''

Define parameters

'''

H_i = 200
rho_i = 910
rho_w = 1000
H_w = rho_i/rho_w * H_i
g = 9.8
eta0 = H_w
d_eta_dt0 = 0
L0 = 5e3
w0 = 2
K_ic = 1e5
sigma0 = -rho_i*g*H_i/2 + rho_w*g/(2*H_i)*eta0**2

# The following calculates Rxx such that Kic=KI, and then adds a tiny perturbation
Rxx = find_min_rxx(L0,sigma0,K_ic)[0]+1



'''

Run simulations

'''

y0 = [eta0, d_eta_dt0,L0]
t = np.linspace(0, 450, 10001)

fig,ax=plt.subplots(3,1,figsize=(10,10))
fig.patch.set_facecolor('w')

for epsilon in (25,40,100,400):

    sol = odeint(water_crack, y0, t, args=(epsilon, H_i, Rxx, w0))

    ax[0].plot(t, sol[:, 0]-H_w)
    ax[0].set_ylabel('Deflection of water surface (m)')
    ax[1].plot(t, np.gradient(sol[:, 2],t), label='L(t), km')
    ax[1].set_ylabel('Crack propagation rate (m/s)')
    ax[2].plot(t, sol[:, 2]/1e3-L0/1e3, label='L(t), km')
    ax[2].set_ylabel('Crack growth (km)')
    print("Mean crack propagation rate: ",np.round(np.mean(np.gradient(sol[:, 2],t)),2)," m/s")
ax[0].grid()
ax[1].grid()
ax[2].grid()
plt.xlabel('time, seconds')
plt.show()
```

```python
'''

Define model

'''

# this version includes nonzero initial width and accounts for alpha's dependence on time
def water_crack(y, t, epsilon, H_i, Rxx, w0):
    
    # get variables and set parameters
    eta, d_eta_dt, L = y
    g = 9.8
    v_r = 2000
    K_c = 1e5
    rho_i = 910
    rho_w = 1000
    H_w = rho_i/rho_w * H_i

    
    # calculate water level
    H_w = rho_i/rho_w * H_i

    # calculate stress and stress intensity factor
    sigma = rho_i*g*H_i/2 - rho_w*g/(2*H_i)*eta**2
    sigma_sum = Rxx - sigma
    K_i = stress_intensity(L,sigma_sum)
    
    # use stress intensity factor to determine if crack propagates
    if K_i < K_c:
        dLdt = 0
    else:
        dLdt = v_r * (1 - (K_c/K_i)**2)

    # use "Eshelby" relation to get width and its time derivative as a function of length
    mu_star = 3.6e9/(1-0.33)
    alpha = sigma_sum / mu_star * np.pi/4
    w = w0 + alpha * L

    # write time derivative of stress sum
    d_sigma_sum_dt = d_eta_dt*eta*rho_w*g/H_i
    
    # write time derivative of alpha
    d_alpha_dt = d_sigma_sum_dt*np.pi/mu_star/4
    
    # write time derivative of width
    dwdt = alpha*dLdt + d_alpha_dt*L
        
    # write time derivative of water volume, V'
    dVdt = w*L*d_eta_dt + w*dLdt*eta + dwdt*L*eta
    
    # write time derivative of stress
    d_sigma_sum_dt = (rho_w*g)/H_i * d_eta_dt * eta
    
    # write 2nd time derivative of crack length
    dLd2t = (v_r*K_c**2)/np.pi * (1/((L**2)*(sigma_sum**3))*(dLdt*sigma_sum + 2*L*d_sigma_sum_dt))
    
    # write expression for terms of V" that do not contain eta"
    gamma = rho_w*g*np.pi/4/H_i/mu_star
    lamb = 2*(w*dLdt*d_eta_dt + dwdt*L*d_eta_dt + dwdt*dLdt*eta) + w*dLd2t*eta + L*eta*(alpha*dLd2t + 2*d_alpha_dt*dLdt + gamma*L*d_eta_dt**2)
    beta = L*(w + gamma*L*eta**2)
    
    # write bernoulli with u' substituted in so all eta" terms can be moved to LHS
    d_eta_d2t = (H_w + epsilon*beta/L)**-1 * (g*(H_w - eta) - (epsilon/L**2)*(lamb*L - dVdt*dLdt))
    
    dydt = [d_eta_dt, d_eta_d2t, dLdt]
    
    return dydt


'''

Define parameters

'''

H_i = 200
rho_i = 910
rho_w = 1000
H_w = rho_i/rho_w * H_i
g = 9.8
eta0 = H_w
d_eta_dt0 = 0
L0 = 5e3
w0 = 2
K_ic = 1e5
sigma0 = -rho_i*g*H_i/2 + rho_w*g/(2*H_i)*eta0**2

# The following calculates Rxx such that K_ic = K_i, and then adds a tiny perturbation
Rxx = find_min_rxx(L0,sigma0,K_ic)[0]+1



'''

Run simulations

'''

y0 = [eta0, d_eta_dt0,L0]
t = np.linspace(0, 450, 10001)

# epsilon = 10e3/400 #"Lc/Hc"

fig,ax=plt.subplots(3,1,figsize=(10,10))
fig.patch.set_facecolor('w')

for epsilon in (25,40,100,400):

    sol = odeint(water_crack, y0, t, args=(epsilon, H_i, Rxx, w0))

    ax[0].plot(t, sol[:, 0]-H_w)
    ax[0].set_ylabel('Deflection of water surface (m)')
    ax[1].plot(t, np.gradient(sol[:, 2],t), label='L(t), km')
    ax[1].set_ylabel('Crack propagation rate (m/s)')
    ax[2].plot(t, sol[:, 2]/1e3, label='L(t), km')
    ax[2].set_ylabel('Crack length (km)')
    print("Mean crack propagation rate (epsilon="+str(epsilon)+"):",np.round(np.mean(np.gradient(sol[:, 2],t)),2)," m/s")
ax[0].grid()
ax[1].grid()
ax[2].grid()
plt.xlabel('time, seconds')
plt.show()
```
