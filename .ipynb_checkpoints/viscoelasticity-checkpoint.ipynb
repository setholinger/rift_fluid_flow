```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''

Find the roots of the viscoelastic dynamic beam equation

'''

# define function to get coefficients 
def get_coefficients(H_i,xi):
    
    # define material properties, elastic moduli, and viscosity (calculated assuming eta = 6 bar*year
    rho_i = 910
    rho_w = 1000
    H_w = H_i*rho_i/rho_w
    g = 9.8
    E = 8.7e9
    nu = 0.3
    lamb = (E*nu)/(1+nu)/(1-2*nu)
    mu = E/2/(1+nu)
    M = lamb + 2*mu
    k = lamb + (2/3)*mu
    eta = 1.5e13
    t_m = mu/eta

    # define parameter groups
    gamma = 1/np.tanh(H_w*xi)/xi;
    B = rho_i*H_i + rho_w*gamma

    # write the coefficients for the polynomial equation- index corresponds to the degree of each monomial term with respect to s 
    c_4 = M*B
    c_3 = B*t_m*(M+k)
    c_2 = B*k*t_m**2 + M*rho_w*g + (1/3)*(H_i**3)*(xi**4)*mu*(mu+lamb)
    c_1 = t_m*(rho_w*g*(M+k) + (1/3)*(H_i**3)*(xi**4)*mu*k)
    c_0 = rho_w*g*k*t_m**2
    coefficients = [c_4,c_3,c_2,c_1,c_0]
    
    return coefficients

# set ice shelf geometry (length is used to lowest wavenumber)
H_i = 200
L = 7e4

# define a range of wavenumbers
xi_vect = np.logspace(np.log10(1/(4*L)),np.log10(1/H_i),100)

# iterate through wavenumber vector and find roots (angular wavenumber/frequency conversion done here!)
roots = []
for xi in xi_vect:
    coefficients = get_coefficients(H_i,2*np.pi*xi)
    roots.append(np.roots(coefficients)/2/np.pi)
roots = np.array(roots)

# plot each root
fig,ax = plt.subplots(4,2,figsize=(15,15))
for i in range(roots.shape[1]):
    
    # plot real roots, multiplying by -1 if negative
    if np.max(roots[:,i].real) > 0:
        ax[i,0].plot(xi_vect,roots[:,i].real)
        ax[i,0].set_ylabel('Frequency (Hz)')
    elif np.max(roots[:,i].real) < 0:
        ax[i,0].plot(xi_vect,-1*roots[:,i].real)
        ax[i,0].set_ylabel('-1 * Frequency (Hz)')
        
    # plot imaginary roots, multiplying by -1 if negative
    if np.max(roots[:,i].imag) > 0 or np.max(roots[:,i].real) == 0:
        ax[i,1].plot(xi_vect,roots[:,i].imag)
        ax[i,1].set_ylabel('Frequency (Hz)')
    elif np.max(roots[:,i].imag) < 0:
        ax[i,1].plot(xi_vect,-1*roots[:,i].imag)
        ax[i,1].set_ylabel('-1 * Frequency (Hz)')
        
    # configure labels
    for j in range(ax.shape[1]):
        ax[i,j].set_xscale('log')
        ax[i,j].set_yscale('log')
        ax[i,j].set_xlabel('Wavenumber ($m^{-1}$)')
        ax[i,j].grid(which='both',color='lightgrey')
    ax[i,0].set_title('Real part of root '+str(i+1))
    ax[i,1].set_title('Imaginary part of root '+str(i+1))
plt.suptitle("Dispersion analysis for viscoelastic beam",size=15)
plt.tight_layout()
```

```python
# plot phase velocities 
fig,ax = plt.subplots(4,2,figsize=(15,15))
for i in range(roots.shape[1]):
    
    # plot real roots, multiplying by -1 if negative
    if np.max(roots[:,i].real) > 0:
        ax[i,0].plot(roots[:,i].imag,roots[:,i].real/xi_vect)
        ax[i,0].set_xlabel('Frequency (Hz)')
        ax[i,0].set_ylabel('Phase velocity (m/s)')
    elif np.max(roots[:,i].real) < 0 and np.max(roots[:,i].imag) > 0:
        ax[i,0].plot(roots[:,i].imag,-1*roots[:,i].real/xi_vect)
        ax[i,0].set_xlabel('Frequency (Hz)')
        ax[i,0].set_xlabel('-1 * Phase velocity (m/s)')
    elif np.max(roots[:,i].real) < 0 and np.max(roots[:,i].imag) < 0:
        ax[i,0].plot(-1*roots[:,i].imag,-1*roots[:,i].real/xi_vect)
        ax[i,0].set_xlabel('-1 * Frequency (Hz)')
        ax[i,0].set_ylabel('-1 * Phase velocity (m/s)')
    elif np.max(roots[:,i].real) < 0 and np.max(roots[:,i].imag) == 0:
        ax[i,0].plot(-1*roots[:,i].real,-1*roots[:,i].real/xi_vect) 
        ax[i,0].set_xlabel('-1 * Frequency (Hz)')
        ax[i,0].set_ylabel('-1 * Phase velocity (m/s)')

    # plot imaginary roots, multiplying by -1 if negative
    if np.max(roots[:,i].imag) > 0 or np.max(roots[:,i].imag) == 0:
        ax[i,1].plot(roots[:,i].imag,roots[:,i].imag/xi_vect)
        ax[i,1].set_xlabel('Frequency (Hz)')
        ax[i,1].set_ylabel('Phase velocity (m/s)')
    elif np.max(roots[:,i].imag) < 0:
        ax[i,1].plot(-1*roots[:,i].imag,-1*roots[:,i].imag/xi_vect)
        ax[i,1].set_xlabel('-1 * Frequency (Hz)')
        ax[i,1].set_ylabel('-1 * Phase velocity (m/s)')

    # configure labels
    for j in range(ax.shape[1]):
        ax[i,j].set_xscale('log')
        ax[i,j].set_yscale('log')
        ax[i,j].grid(which='both',color='lightgrey')
    ax[i,0].set_title('Real part of root '+str(i+1))
    ax[i,1].set_title('Imaginary part of root '+str(i+1))
plt.suptitle("Velocity analysis for viscoelastic beam",size=15)
plt.tight_layout()
```

```python
# get complex-valued roots
complex_roots = roots[:,np.iscomplex(roots[0,:])]

# plot phase velocities 
fig,ax = plt.subplots(2,1,figsize=(15,15))
for i in range(complex_roots.shape[1]):
    
    # plot quality factor Q
    if np.max(complex_roots[:,i].imag) < 0:
        ax[i].plot(-1*complex_roots[:,i].imag,np.abs(complex_roots[:,i].imag/complex_roots[:,i].real/2))
        ax[i].set_xlabel('-1 * Frequency (Hz)')
        ax[i].set_ylabel('Quality factor Q')
    else:
        ax[i].plot(complex_roots[:,i].imag,np.abs(complex_roots[:,i].imag/complex_roots[:,i].real/2))
        ax[i].set_xlabel('Frequency (Hz)')
        ax[i].set_ylabel('Quality factor Q')

    # configure labels
    ax[i].set_xscale('log')
    ax[i].set_yscale('log')
    ax[i].grid(which='both',color='lightgrey')
    ax[i].set_title('Q for root '+str(i+1))
plt.suptitle("Attenuation analysis for viscoelastic beam",size=15)
plt.tight_layout()
```

```python
'''

Use dispersion relation to calculate resonant frequencies

'''

def dispersion(xi,H_i):
    g = 9.8
    rho_i = 910
    rho_w = 1024
    H_w = H_i * rho_i / rho_w 
    E = 8.7e9
    nu = 0.3
    D = (E * H_i**3) / (12*(1-nu**2))
    gamma = 1/np.tanh(H_w*xi)/xi;
    omega = np.sqrt((D*xi**4 + rho_w*g)/(rho_i*H_i + rho_w*gamma))
    return omega 

def resonance(L,H_i,mode):
    xi = (mode+1)*np.pi/L
    omega = dispersion(xi,H_i)
    t = 2*np.pi/omega
    return t

# set parameters
L = 30000
H_i = 750

# first 3 modes
mode_1_period = resonance(L,H_i,0)
mode_2_period = resonance(L,H_i,1)
mode_3_period = resonance(L,H_i,2)
print(mode_1_period)
print(mode_2_period)
print(mode_3_period)
```
