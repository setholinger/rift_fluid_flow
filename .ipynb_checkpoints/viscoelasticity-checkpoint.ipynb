```python
'''

Define function to find the roots of the viscoelastic dynamic beam equation with numpy.roots

'''



# define function to get coefficients 
def get_coefficients(H_i,H_w,xi,eta,physics,dimensional=0):
    
    # decide which physics to include
    flexure, water_waves, inertia, buoyancy = physics
    
    # define material properties, elastic moduli, and viscosity (calculated assuming eta = 6 bar*year)
    rho_i = 910
    rho_w = 1000
    g = 9.8
    E = 8.7e9
    nu = 0.3
    mu = E/2/(1+nu)
    t_m = eta/mu
    nu_1 = 3*(1-nu)/(1+nu)
    nu_2 = 3/(2*(1+nu))
    
    if dimensional:
        # inertia
        if inertia:
            N = np.sqrt(rho_w*g/(rho_i*H_i))
            I = 1/(N**2)
        else:
            I = 0

        # water waves
        if water_waves:
            #gamma = 1/np.tanh(H_w*xi)/xi
            gamma = 1/(H_w*xi**2)
            W = gamma/g
        else: 
            W = 0
    else:
        # inertia
        if inertia:
            N = np.sqrt(rho_w*g/(rho_i*H_i))
            I = 1/(t_m**2 * N**2)
        else:
            I = 0

        # water waves
        if water_waves:
            #gamma = 1/np.tanh(H_w*xi)/xi/(t_m**2)
            gamma = 1/(H_w*xi**2)/(t_m**2)
            W = gamma/g
        else: 
            W = 0
        
    # flexure
    if flexure:
        F = (1/3 * H_i**3 * mu * xi**4)/(rho_w*g)
    else:
        F = 0
        
    # buoyancy
    if buoyancy:
        B = 1
    else:
        B = 0
        
    # write the coefficients for the polynomial equation- index corresponds to the degree of each monomial term with respect to s 
    c_4 = (I+W) * nu_1
    c_3 = (I+W) * (nu_1+1)
    c_2 = (I+W) + B*nu_1 + F*nu_2
    c_1 = (B*(nu_1+1) + F)
    c_0 = B
    coefficients = [c_4,c_3,c_2,c_1,c_0]
    
    return coefficients
```

```python
'''

Define a few functions to plot the results

'''

def dispersion_plot(xi_vect,roots,eta_vect):

    # elastic parameters
    E = 8.7e9
    nu = 0.3
    mu = E/2/(1+nu)
    
    
    # plot each root
    fig,ax = plt.subplots(4,2,figsize=(15,15))
    for i in range(roots.shape[1]):
        for j in range(roots.shape[0]):

            # plot real roots, multiplying by -1 if negative
            if np.max(roots[j,i,:].real) > 0:
                ax[i,0].plot(xi_vect,roots[j,i,:].real,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,0].set_ylabel('Frequency ($s*t_m$)')
                ax[i,0].axhline(y=1,linestyle='--')
            elif np.max(roots[j,i,:].real) < 0:
                ax[i,0].plot(xi_vect,-1*roots[j,i,:].real,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,0].set_ylabel('-1 * Frequency ($s$)')
                ax[i,0].axhline(y=1,linestyle='--')

            # plot imaginary roots, multiplying by -1 if negative
            if np.max(roots[j,i,:].imag) > 0 or np.max(roots[j,i,:].real) == 0:
                ax[i,1].plot(xi_vect,roots[j,i,:].imag,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,1].set_ylabel('Frequency ($s*t_m$)')
            elif np.max(roots[j,i,:].imag) < 0:
                ax[i,1].plot(xi_vect,-1*roots[j,i,:].imag,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,1].set_ylabel('-1 * Frequency ($s*t_m$)')
            elif np.max(roots[j,i,:].imag) == 0:
                #ax[i,0].axhline(y=mu/eta_vect[j],linestyle='--',c="C1")
                ax[i,0].axhline(y=1,linestyle='--')

            # configure labels
            for k in range(ax.shape[1]):
                ax[i,k].set_xscale('log')
                ax[i,k].set_yscale('log')
                ax[i,k].set_xlabel('Wavenumber ($m^{-1}$)')
                ax[i,k].grid(which='both',color='lightgrey')
                ax[i,k].legend()
                
            #ax[i,1].axhline(y=mu/eta_vect[j],linestyle='--',c="C1")
            ax[i,1].axhline(y=1,linestyle='--')
            ax[i,0].set_title('Real part of root '+str(i+1))
            ax[i,1].set_title('Imaginary part of root '+str(i+1))
    plt.suptitle("Dispersion analysis for viscoelastic beam",size=15)
    plt.tight_layout()
    
    
    
def velocity_plot(xi_vect,roots,eta_vect):
    
    # elastic parameters
    E = 8.7e9
    nu = 0.3
    mu = E/2/(1+nu)
    
    # plot each root
    fig,ax = plt.subplots(4,2,figsize=(15,15))
    for i in range(roots.shape[1]):
        for j in range(roots.shape[0]):    

            # plot real roots, multiplying by -1 if negative
            if np.max(roots[j,i,:].real) > 0:
                ax[i,0].plot(roots[j,i,:].imag,roots[j,i,:].real/xi_vect,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,0].set_xlabel('Frequency ($s*t_m$)')
                ax[i,0].set_ylabel('Phase velocity (m/s)')
            elif np.max(roots[j,i,:].real) < 0 and np.max(roots[j,i,:].imag) > 0:
                ax[i,0].plot(roots[j,i,:].imag,-1*roots[j,i,:].real/xi_vect,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,0].set_xlabel('Frequency ($s*t_m$)')
                ax[i,0].set_ylabel('-1 * Phase velocity (m/s)')
            elif np.max(roots[j,i,:].real) < 0 and np.max(roots[j,i,:].imag) < 0:
                ax[i,0].plot(-1*roots[j,i,:].imag,-1*roots[j,i,:].real/xi_vect,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,0].set_xlabel('-1 * Frequency ($s*t_m$)')
                ax[i,0].set_ylabel('-1 * Phase velocity (m/s)')
            elif np.max(roots[j,i,:].real) < 0 and np.max(roots[j,i,:].imag) == 0:
                ax[i,0].plot(-1*roots[j,i,:].real,-1*roots[j,i,:].real/xi_vect,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,0].set_xlabel('-1 * Frequency ($s*t_m$)')
                ax[i,0].set_ylabel('-1 * Phase velocity (m/s)')
            ax[i,0].axvline(x=1,linestyle='--')

            # plot imaginary roots, multiplying by -1 if negative
            if np.max(roots[j,i,:].imag) > 0 or np.max(roots[j,i,:].imag) == 0:
                ax[i,1].plot(roots[j,i,:].imag,roots[j,i,:].imag/xi_vect,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,1].set_xlabel('Frequency ($s*t_m$)')
                ax[i,1].set_ylabel('Phase velocity (m/s)')
            elif np.max(roots[j,i,:].imag) < 0:
                ax[i,1].plot(-1*roots[j,i,:].imag,-1*roots[j,i,:].imag/xi_vect,label="$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$")
                ax[i,1].set_xlabel('-1 * Frequency ($s*t_m$)')
                ax[i,1].set_ylabel('-1 * Phase velocity (m/s)')
            ax[i,1].axvline(x=1,linestyle='--')

            # configure labels
            for k in range(ax.shape[1]):
                ax[i,k].set_xscale('log')
                ax[i,k].set_yscale('log')
                ax[i,k].grid(which='both',color='lightgrey')
                ax[i,k].legend()
                #ax[i,k].axvline(x=mu/eta_vect[j],linestyle='--',c="C1")
                ax[i,k].axvline(x=1,linestyle='--')
            ax[i,0].set_title('Real part of root '+str(i+1))
            ax[i,1].set_title('Imaginary part of root '+str(i+1))
    plt.suptitle("Velocity analysis for viscoelastic beam",size=15)
    plt.tight_layout()
    
    

def combined_velocity_plot(xi_vect,roots,eta_vect):
    
    # elastic parameters
    E = 8.7e9
    nu = 0.3
    mu = E/2/(1+nu)
    
    # plot each root
    fig,ax = plt.subplots(1,1,figsize=(15,10))
    for i in range(roots.shape[1]):
        for j in range(roots.shape[0]):    

            # get label for legend
            label = "$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$"
            
            # plot real roots, multiplying by -1 if negative
            if np.max(roots[j,i,:].real) > 0:
                ax.plot(roots[j,i,:].imag,roots[j,i,:].real/xi_vect,label=label)
                ax.set_xlabel('Frequency ($s*t_m$)')
            elif np.max(roots[j,i,:].real) < 0 and np.max(roots[j,i,:].imag) > 0:
                ax.plot(roots[j,i,:].imag,-1*roots[j,i,:].real/xi_vect,label=label)
                ax.set_xlabel('Frequency ($s*t_m$)')
            elif np.max(roots[j,i,:].real) < 0 and np.max(roots[j,i,:].imag) < 0:
                ax.plot(-1*roots[j,i,:].imag,-1*roots[j,i,:].real/xi_vect,label=label)
                ax.set_xlabel('-1 * Frequency ($s*t_m$)')
            elif np.max(roots[j,i,:].real) < 0 and np.max(roots[j,i,:].imag) == 0:
                ax.plot(-1*roots[j,i,:].real,-1*roots[j,i,:].real/xi_vect,label=label)
                ax.set_xlabel('-1 * Frequency ($s*t_m$)')

            # plot imaginary roots, multiplying by -1 if negative
            if np.max(roots[j,i,:].imag) > 0 or np.max(roots[j,i,:].imag) == 0:
                ax.plot(roots[j,i,:].imag,roots[j,i,:].imag/xi_vect,label=label,c="k")
                ax.set_xlabel('Frequency ($s*t_m$)')
            elif np.max(roots[j,i,:].imag) < 0:
                ax.plot(-1*roots[j,i,:].imag,-1*roots[j,i,:].imag/xi_vect,label=label,c="k")
                ax.set_xlabel('-1 * Frequency ($s*t_m$)')

            # configure labels
            ax.set_ylabel('s/k')
            ax.set_xscale('log')
            ax.set_yscale('log')
            ax.grid(which='both',color='lightgrey')
            #ax.axvline(x=mu/eta_vect[j],linestyle='--',c='C1')
            ax.axvline(x=1,linestyle='--')
    plt.suptitle("Velocity analysis for viscoelastic beam",size=15)
    plt.tight_layout()
    
    
    
def decay_plot(xi_vect,roots,eta_vect):
    
    # elastic parameters
    E = 8.7e9
    nu = 0.3
    mu = E/2/(1+nu)
    
    # plot each root 
    fig,ax = plt.subplots(figsize=(15,10))
    for j in range(roots.shape[0]):    

        # get label for legend
        label = "$\eta=10^{"+str(np.round(np.log10(eta_vect[j]),2))+"}$"

        # plot quality factor Q
        if np.max(roots[j,0,:].imag) < 0:
            ax.plot(-1*roots[j,0,:].imag,np.abs(roots[j,0,:].imag/roots[j,0,:].real/2),label=label)
            ax.set_xlabel('Frequency ($s*t_m$)')
            #ax.plot(xi_vect,np.abs(roots[j,0,:].imag/roots[j,0,:].real/2),label=label)
            #ax.set_xlabel('Wavenumner ($xi$)')
            ax.set_ylabel('Quality factor Q')
        else:
            ax.plot(roots[j,0,:].imag,np.abs(roots[j,0,:].imag/roots[j,0,:].real/2),label=label)
            ax.set_xlabel('Frequency ($s*t_m$)')
            #ax.plot(xi_vect,np.abs(roots[j,0,:].imag/roots[j,0,:].real/2),label=label)
            #ax.set_xlabel('Wavenumner ($xi$)')
            ax.set_ylabel('Quality factor Q')

        # configure labels
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.grid(which='both',color='lightgrey')
        ax.set_title('Quality factor Q')
        ax.legend()
        #ax.axvline(x=mu/eta_vect[j],linestyle='--',c="C1")
        ax.axvline(x=1,linestyle='--')
    
    # plot a line to compare slopes
    #f = np.array([0,0.1,0.001,0.0001,0.00001])
    #ax.plot(f,1e2*(f)**(-3))
    
    plt.suptitle("Attenuation analysis for viscoelastic beam",size=15)
    plt.tight_layout()
    
    
    
def decay_comparison_plot(xi_vect,roots,approximate_solution,eta,period=np.nan):
    
    # elastic parameters
    E = 8.7e9
    nu = 0.3
    mu = E/2/(1+nu)
    t_m = eta/mu
    
    # plot each root 
    fig,ax = plt.subplots(figsize=(15,10))
    
    # plot quality factor Q
    Q = np.abs(roots[0,0,:].imag/roots[0,0,:].real/2)
    if np.max(roots[0,0,:].imag) < 0:
        ax.plot(-1*roots[0,0,:].imag,Q,label="$Q$")
        ax.set_xlabel('Nondimensional Frequency ($st_m$)')
        #ax.plot(xi_vect,np.abs(roots[0,0,:].imag/roots[j,0,:].real/2),label="Q")
        #ax.set_xlabel('Wavenumner ($xi$)')
        ax.set_ylabel('Quality factor Q')
    else:
        ax.plot(roots[0,0,:].imag,Q,label="$Q$")
        ax.set_xlabel('Nondimensional Frequency ($st_m$)')
        #ax.plot(xi_vect,np.abs(roots[0,0,:].imag/roots[j,0,:].real/2),label="Q")
        #ax.set_xlabel('Wavenumner ($xi$)')
        ax.set_ylabel('Quality factor Q')

    # plot approximate solution
    approximate_Q = abs(approximate_solution.imag/approximate_solution.real)/2
    ax.plot(approximate_solution.imag,approximate_Q,'k',linestyle='--',label="$Q_{asym}$")
    #ax.plot(xi_vect,abs(np.array(approximate_solution).imag/np.array(approximate_solution).real)/2)

    # configure labels
    plt.rcParams.update({'font.size': 15})
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.grid(which='both',color='lightgrey')
    ax.legend()
    
    # plot line at maxwell time and period of interest
    ax.axvline(x=1,c='r',linestyle='--')
    ax.text(0.8,min(Q)/2400,'$t_m$',fontsize=15,c='r')
    ax.axvline(x=t_m/period,c='g',linestyle='--')
    ax.text(t_m/period-100,min(Q)/2400,'T = '+str(period)+'s',fontsize=15,c='g')
    
    # label Q value for desired period
    period_Q = Q[np.argmin(abs(np.abs(roots[0,0,:].imag)-t_m/period))]
    power = np.floor(np.log10(period_Q))
    num = np.round(period_Q/(10**power),2)
    ax.text(t_m/period,period_Q*2,'  Q = '+str(num)+"e"+str(int(power)),fontsize=15,c="C0")
    
    # label approximate Q value for desired period
    approx_period_Q = approximate_Q[np.argmin(abs(np.abs(approximate_solution.imag)-t_m/period))]
    power = np.floor(np.log10(approx_period_Q))
    num = np.round(approx_period_Q/(10**power),2)
    ax.text(t_m/period,approx_period_Q,'  $Q_{asym}$ = '+str(num)+"e"+str(int(power)),fontsize=15,c='k')

    ax.set_ylim(min(Q)/1000,max(Q))

    power = np.floor(np.log10(eta))
    num = int(eta/(10**power)) if (eta/(10**power)).is_integer() else eta/(10**power)
    eta_string = "$\eta=" + str(num)+"e"+str(int(power))+"$"
    t_m_string = "$t_m=$"+str(np.round(eta/mu/60/60,2))+" hours"
    plt.suptitle("Attenuation analysis for viscoelastic beam ("+eta_string+", " + t_m_string+")",size=15)
    plt.tight_layout()
```

```python
'''

Find roots of the viscoelastic dynamic beam equation with numpy.roots

'''

# set ice shelf geometry (length is used to set lowest wavenumber)
H_i = 550
H_w = 700
L = 7e4

# define a range of wavenumbers
xi_vect = np.logspace(-5,np.log10(1000/H_i),100)

# define a range of viscosities
eta_vect = [3e14]

# choose which physics to include
flexure = 1
water_waves = 1
inertia = 1
buoyancy = 1
physics = [flexure,water_waves,inertia,buoyancy]

# iterate through wavenumber vector and find roots
roots = []
for eta in eta_vect:
    roots_eta = []
    for xi in xi_vect:
        coefficients = get_coefficients(H_i,H_w,xi/2/np.pi,eta,physics)
        roots_eta.append(np.roots(coefficients))
    roots.append(np.transpose(roots_eta))
roots = np.array(roots)

# make plots to display results
dispersion_plot(xi_vect,roots,eta_vect)
velocity_plot(xi_vect,roots,eta_vect)
combined_velocity_plot(xi_vect,roots,eta_vect)
decay_plot(xi_vect,roots,eta_vect)
```
